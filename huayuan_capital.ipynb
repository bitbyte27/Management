### 华园控股控参股子公司-实缴注资情况（更新至2018-11-27）

```python
import pandas as pd
import numpy as np
```

```python
Subsidiary_List = pd.read_excel('csv/capital.xlsx', sheetname = 'capital', encoding='gbk')
```

```python
def capital(company):
    SL = Subsidiary_List.loc[Subsidiary_List[u'公司简称'] == company].copy()
    SL.iloc[:,5] = SL[u'持股比例'].map(lambda x:format(x,'.2%'))
    for i in range(6,13):
        SL.iloc[:,i] = SL.iloc[:,i].map(lambda x:format(x,','))
    SL = SL.iloc[:,2:].copy().replace('nan', '')
    return SL
```

```python
SL_Y_name = Subsidiary_List.loc[Subsidiary_List[u'类别'] == u'医疗集团',[u'公司简称']].copy()
SL_Y_name.drop_duplicates()
```
